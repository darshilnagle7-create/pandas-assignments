# Q1. Display Unique Values

```python
City = ["Bhopal", "Indore", "Bhopal", "Delhi"]
```

## Sample Output


# Q4. Group By City

## Problem Statement
Find average marks city-wise using groupby().

## Sample Input
```python
City   Marks
Bhopal 80
Bhopal 90
Delhi  70
```

## Sample Output
```python
Bhopal 85
Delhi  70
```


````python
# Write your code here
a = {
    'City':["Bhopal", "Indore", "Bhopal", "Delhi"],
    'Marks':[80,90,70,60]
}
b = pd.DataFrame(a)
b
city_marks = b.groupby('City')['Marks'].mean()
print(city_marks)
````

```python
Filtered rows
```


# Q9. Group By Department

## Problem Statement
Create Employee and Department data and find total salary department-wise.

## Sample Output
```python
HR       50000

```


````python
# Write your code here
a = {
    "Employee": ["Aman", "Riya", "Rohit", "Priya"],
    "Department": ["HR", "IT", "HR", "IT"],
    "Salary": [20000, 50000, 30000, 70000]
}
b = pd.DataFrame(a)
b
b.groupby('Department')['Salary'].sum()
````


# Q10. Complete Practice Question

## Problem Statement
Create a DataFrame containing:
- Name
- Age
- City
- Marks

Perform:
1. Show unique cities
2. Count city frequency
3. Filter marks > 80
4. Sort by marks

## Sample Output
```python
Required Results
```


````python
# Write your code here
a = {
    "Name": ["Aman", "Riya", "Rohit", "Priya"],
    "Age": [20, 21, 22, 20],
    "City": ["Bhopal", "Indore", "Bhopal", "Delhi"],
    "Marks": [85, 92, 78, 88]
}
b = pd.DataFrame(a)
b
b['City'].unique()
b['City'].value_counts()
b[b['Marks'] > 80]
b.sort_values(by = 'Marks')
````
