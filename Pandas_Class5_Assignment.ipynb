## Q1. Sort Age in Ascending Order

Input:
Age = [25, 18, 30, 22]

Output:
18
22
25
30

```python
# Write your code here
import pandas as pd
a = ({
    'Age':[25,18,30,22]
})
b = pd.DataFrame(a)
b
b.sort_values(by='Age')
```

## Q15. Mini Practice

Create a DataFrame with:
Name, Age, Score, Height

Tasks:
1. Sort by Score (Descending)
2. Find Mean Age
3. Find Median Height
4. Show Score Frequency

```python
# Write your code here
a = ({
    'Name': ['Aman', 'Riya', 'Karan', 'Neha', 'Rohit'],
    'Age': [20, 21, 22, 19, 23],
    'Score': [85, 92, 78, 88, 95],
    'Height': [170, 165, 175, 160, 180]
})
b = pd.DataFrame(a)
b
b.sort_values(by='Score',ascending=False)
b['Age'].mean()
b['Height'].median()
b.value_counts('Score')
```
